```python
#Monte carlo's simualtion technique in order to estimate different values for PI


import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
from numpy.random import random
from numpy import pi

def Pi_MonteCarloSimulation(No_of_points):
    circle_x,circle_y = [],[]#To draw a circle for simulation
    squareX = [1,-1,-1,1,1]#To draw a square 
    squareY = [1,1,-1,-1,1]
    
    #Construction of the ponits
    for i in range(361):
        circle_x.append(np.cos(np.pi*i/180))
        circle_y.append(np.sin(np.pi*i/180))

    insideCircle_x,insideCircle_y,outsideCircle_x,outsideCircle_y,Iterations,Instant_Pi = [],[],[],[],[],[]
    Counter = 0

#Random points genereated in estimation of pi value
    for i in range(No_of_points):
        x = 2*(random()-0.5)
        y = 2*(random()-0.5)
        r = np.sqrt(x**2+y**2)
        Iterations.append(i)
        if r <= 1:
            Counter +=1
            insideCircle_x.append(x)
            insideCircle_y.append(y)
        else:
            outsideCircle_x.append(x)
            outsideCircle_y.append(y)
        Instant_Pi.append(4*Counter/(i+1))

    piValue = 4*Counter / No_of_points
    piError = round(100*((piValue-pi)/pi),4)

    rcParams['figure.figsize'] = 5, 5
    plt.plot(squareX,squareY,color='#000000')
    plt.plot(circle_x,circle_y,color='#0000CC')
    plt.scatter(insideCircle_x,insideCircle_y,color='#fdc800',marker=".")
    plt.scatter(outsideCircle_x,outsideCircle_y,color='#CC0000',marker=".")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    #Comparison of accuracy of estimation with line chart
    plt.plot(Iterations,Instant_Pi,color='#009900')
    plt.axhline(y=pi,color='#0F0F0F',ls='--')#for the acutal generic value 3.14
    plt.axis([0,No_of_points,0,4.1])
    plt.xlabel('Number of points')
    plt.ylabel('Estimated pi values')
    plt.show()
Pi_MonteCarloSimulation(100)
```
